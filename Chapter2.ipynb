## 2.1 wave function

$\frac{\partial q}{\partial t} + c \frac{\partial q}{\partial x} = 0 (c>0)$

$q = f(x-ct)$

## 2.2 Solution of wave function

### FTCS method

Forward in time

$\frac{\partial q}{\partial t} = \frac{q_j^{n+1} - q_j^{n}}{\Delta t}$

Centered in space

$\frac{\partial q}{\partial x} = \frac{q_{j+1}^n - q_{j-1}^n}{2\Delta x}$

### Programming and figure 1

```python
#Initialization of const number

c = 1
dt = 0.05
dx = 0.1
jmax = 21
nmax = 6
x = np.linspace(0,dx*(jmax -1),jmax)

q = np.zeros(jmax)
for j in range(jmax):
    if (j < jmax /2):
        q[j] = 1
    else:
        q[j] = 0
```

```python
plt.figure(figsize=(7,7),dpi=100)
plt.rcParams["font.size"] = 22

plt.plot(x,q,marker="o",lw=2,label="n=0")

for n in range(1,nmax+1):
    qold =q.copy()
    for j in range(1,jmax-1):
        q[j] = qold[j]-dt *c*(qold[j+1]-qold[j-1])/(2 * dx)
    if n % 2 ==0:
        plt.plot(x,q,marker="o",lw=2,label=f'n={n}')

plt.grid(color="black",linestyle="dashed",linewidth=0.5)
plt.xlim([0,2.0])
plt.xlabel("x")
plt.ylabel("q")
plt.legend()
plt.show()
```

### Lax Method

Main method: Average first part

$q_j^{n+1} = 1/2 (q_{j+1}^n +q_{j-1}^n) - c/2 \frac{\Delta t}{\Delta x} (q_{j+1}^n-q_{j-1}^n)$

### Programming and figure 4

```python
#Initialization of const number

c = 1
dt = 0.05
dx = 0.1
jmax = 21
nmax = 6
x = np.linspace(0,dx*(jmax -1),jmax)

q = np.zeros(jmax)
for j in range(jmax):
    if (j < jmax /2):
        q[j] = 1
    else:
        q[j] = 0
```

```python
plt.figure(figsize=(7,7),dpi=100)
plt.rcParams["font.size"] = 22

plt.plot(x,q,marker="o",lw=2,label="n=0")

for n in range(1,nmax+1):
    qold =q.copy()
    for j in range(1,jmax-1):
        q[j] = 0.5 * (qold[j+1]+qold[j-1]) - dt *c*(qold[j+1]-qold[j-1])/(2 * dx)
    if n % 2 ==0:
        plt.plot(x,q,marker="o",lw=2,label=f'n={n}')

plt.grid(color="black",linestyle="dashed",linewidth=0.5)
plt.xlim([0,2.0])
plt.xlabel("x")
plt.ylabel("q")
plt.legend()
plt.show()
```

### Lax-Wendroff Method

$q_j^{n+1} = q_j^n - \frac{c}{2} \frac{\Delta t}{\Delta x}(q_{j+1}^n +q_{j-1}^n) + \frac{c^2}{2} \frac{\Delta t}{\Delta x} (q_{j+1}^n-2q_j^n + q_{j-1}^n)$

### Programming and figure 5

```python
#Initialization of const number

c = 1
dt = 0.05
dx = 0.1
jmax = 21
nmax = 6
x = np.linspace(0,dx*(jmax -1),jmax)

q = np.zeros(jmax)
for j in range(jmax):
    if (j < jmax /2):
        q[j] = 1
    else:
        q[j] = 0
```

```python
plt.figure(figsize=(7,7),dpi=100)
plt.rcParams["font.size"] = 22

plt.plot(x,q,marker="o",lw=2,label="n=0")

for n in range(1,nmax+1):
    qold =q.copy()
    for j in range(1,jmax-1):
        q[j] = qold[j] - c / 2 * dt / dx * (qold[j+1]-qold[j-1]) + c**2 /2 * dt / dx * (q[j+1] - 2 * q[j] + q[j-1])
    if n % 2 ==0:
        plt.plot(x,q,marker="o",lw=2,label=f'n={n}')

plt.grid(color="black",linestyle="dashed",linewidth=0.5)
plt.xlabel("x")
plt.ylabel("q")
plt.legend()
plt.title("Lax-Wendroff Result")
plt.show()
```

## 2.3 Linear problems with an undetermined sign of the advection velocity

c > 0 :

$q_j^{n+1} = q_j^n - \Delta \cdot c \frac{q_j^n - q_{j-1}^n}{\Delta x}$

c < 0:

$q_j^{n+1} = q_j^n - \Delta \cdot c \frac{q_{j+1}^n - q_{j}^n}{\Delta x}$

```python
def Calculate(x,q,c,dt,nmax,dx):
    plt.figure(figsize=(7,7),dpi=100)
    plt.rcParams["font.size"] = 22
    plt.plot(x,q,marker="o",lw=2,label="n=0")

    for n in range (1, nmax +1):
        qold = q.copy()
        for j in range(1,jmax -1):
            c2 = 0.5 * (c + abs(c))
            c3 = 0.5 * (c -abs(c))
            q[j] = qold[j] - dt * ((c2 * (qold[j]-qold[j-1])/dx) + (c3 * (qold[j+1]-qold[j])/dx))
        
        if n %2 ==0:
            plt.plot(x,q,marker="o",lw=2,label=f'n={n}')
    plt.grid(color="black",linestyle="dashed",linewidth=0.5)
    plt.xlabel("x")
    plt.ylabel("q")
    plt.legend()
    plt.title("Upwind Result")
    plt.show()
```

```python
c = 1
dt = 0.05
dx = 0.1
jmax = 21
nmax = 6
x = np.linspace(0,dx*(jmax -1),jmax)

q = np.zeros(jmax)
for j in range(jmax):
    if (j < jmax /2):
        q[j] = 1
    else:
        q[j] = 0
```

```python
Calculate(x,q,c,dt,nmax,dx)
```

## 2.4 Value Flux

We start from :

$\frac{\partial q}{\partial t} + \frac{\partial f}{\partial x} = 0$

In fluid mechanics, $f$ is referred to as the flux.

Thus :

$q_{j}^{n+1} = q_{j}^{n} - \frac{\Delta t}{\Delta x} (\hat{f}_{j + \frac{1}{1}}^n -\hat{f}_{j - \frac{1}{1}}^n)$

#### Review of Scalar Advection Equation in FLUX

FTCS method :

$\hat{f}_{j+\frac{1}{2}}^n = \frac{\hat{f}_{j+1}^n + \hat{f}_{j}^n}{2}$

Upwind scheme

$\hat{f}_{j+\frac{1}{2}}^n =\hat{f}_{j}^n$

Lax method:

$\hat{f}_{j+\frac{1}{2}}^n =\frac{1}{2} \left((1-\frac{1}{v})\hat{f}_{j+1}^n + (1+\frac{1}{v})\hat{f}_{j}^n \right)$

where $v = c\cdot \frac{\Delta t}{\Delta x}$

Lax-Wendroff method

$\hat{f}_{j+\frac{1}{2}}^n =\frac{1}{2} \left((1-v)\hat{f}_{j+1}^n + (1+v)\hat{f}_{j}^n \right)$

#### Reconstruct by flux

```python
dt = 0.05
dx = 0.1
jmax = 21
nmax = 6
```

```python
def init(q1,q2,dx,jmax):
    x = np.linspace(0,dx * (jmax -1),jmax)
    q = np.array([(float(q1) if i < 0.5 * jmax else float(q2) ) for i in range (jmax)])
    return (x,q)
```

```python
# FTCS
def FTCS(q,c,dt,dx,j):
    return 0.5 * c * (q[j+1]+q[j])
# Upwind1
def upwind(q,c,dt,dx,j):
    return c * q[j]
# Lax
def lax(q,c,dt,dx,j):
    nv = 1 / (c * dt / dx)
    return 0.5 * c * ((1-nv) * q[j+1] + (1+nv) * q[j])
# LW
def laxwen(q,c,dt,dx,j):
    nv1 = c * dt / dx
    return 0.5 * c * ((1-nv1) * q[j+1] + (1+nv1) * q[j])
```

```python
def do_compute(x,q,c,dt,nmax,dx,ff):
    plt.figure(figsize=(7,7),dpi=100)
    plt.rcParams["font.size"] = 22
    plt.plot(x,q,marker="o",lw=2,label="n=0")

    for n in range (1, 7):
        qold = q.copy()
        for j in range(1,jmax -1):
            ff1 = ff(qold,c,dt,dx,j)
            ff2 = ff(qold,c,dt,dx,j-1)
            q[j] = qold[j] - dt / dx * (ff1-ff2)
        
        if n %2 ==0:
            plt.plot(x,q,marker="o",lw=2,label=f'n={n}')
    plt.grid(color="black",linestyle="dashed",linewidth=0.5)
    plt.xlim([0,2])
    plt.xlabel("x")
    plt.ylabel("q")
    plt.legend()
    plt.title(f"{ff.__name__}")
    plt.show()
```

```python
c = 1
q1, q2 = 1, 0
x,q = init(q1,q2,dx,jmax)
```

```python
do_compute(x,q,c,dt,dx,nmax,FTCS)
```

```python
do_compute(x,q,c,dt,dx,nmax,upwind)
```

```python
do_compute(x,q,c,dt,dx,nmax,lax)
```

```python
do_compute(x,q,c,dt,dx,nmax,laxwen)
```
