## 1.1 Discretization of the trajectory of a moving sphere.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Initialization of const number

g = 9.8
v0 = 10
h0 = 0
dt = 0.05
```

```python
#figure config
plt.figure(figsize=(20,20),dpi=150)
plt.rcParams["font.size"]=20
```

```python
# Analytical Solution

t = np.linspace(0, 2 * v0 / g, 100)
h = -0.5 * g * t**2 + v0 * t + h0
La, = plt.plot(t,h,color="blue")


# Numerical Solution
t = 0
h = h0
while h >= h0:
    Ln = plt.scatter(t,h,marker="o",c="black")
    h += (-g * t+v0) * dt
    t += dt

plt.grid(color="black",linestyle="dashed",linewidth=0.5)
plt.xlabel("Time")
plt.ylabel("Height")
plt.legend(handles=[La,Ln],labels = ["Analytical","Numerical"])
plt.show()
```

## 1.2 Taylor expansion (Taylor series expansion)

```python
import pylatexenc
```

$f(x) = f(a)+ \frac{f^1(a)}{1!}(x-a) + \frac{f^{(2)}(a)}{2!}(x-a)^2 ...\frac{f^{(n)}(a)}{n!}(x-a)^n$

Assume:

$x+\Delta x \rightarrow x$

$x \rightarrow a$

$f(x + \Delta x) = f(x) + \frac{f^{(1)}(x)}{1!}(\Delta x) + \frac{f^{(2)}(x)}{2!}(\Delta x)^2 +... + \frac{f^{(n)}(x)}{n!}(\Delta x)^n...$

According to Mean Value Theorem:

$q_{j+1} = q_j +\Delta x\frac{\partial q}{\partial x} |_j + \frac{1}{2}(\Delta x)^2 \frac{\partial^2 q}{\partial x^2} |_j + \frac{1}{6}(\Delta x)^3 \frac{\partial^3 q}{\partial x^3} |_j +...$

$q_{j-1} = q_j -\Delta x\frac{\partial q}{\partial x} |_j + \frac{1}{2}(\Delta x)^2 \frac{\partial^2 q}{\partial x^2} |_j - \frac{1}{6}(\Delta x)^3 \frac{\partial^3 q}{\partial x^3} |_j +...$

$q_{j+2} = q_j + 2\Delta x\frac{\partial q}{\partial x} |_j + \frac{1}{2}(2\Delta x)^2 \frac{\partial^2 q}{\partial x^2} |_j + \frac{1}{6}(2\Delta x)^3 \frac{\partial^3 q}{\partial x^3} |_j +...$

$q_{j-2} = q_j - 2\Delta x\frac{\partial q}{\partial x} |_j + \frac{1}{2}(2\Delta x)^2 \frac{\partial^2 q}{\partial x^2} |_j - \frac{1}{6}(2\Delta x)^3 \frac{\partial^3 q}{\partial x^3} |_j +...$

### 中心差分 central difference

$\frac{\partial q}{\partial x} |_j = \frac{q_{j+1} - q_{j-1}}{2\Delta X}+ O(x^2)$

### 前进差分 forward difference

$\frac{\partial q}{\partial x} |_j = \frac{q_{j+1} - q_{j}}{\Delta X}+ O(x)$

### 后退差分 backward difference

$\frac{\partial q}{\partial x} |_j = \frac{q_{j} - q_{j-1}}{\Delta X}+ O(x)$
